```python
f = open("张爱玲_倾城之恋.txt", "r")
novel_text = f.read()
f.close()
words = [word.text for word in nlp(novel_text) if word.pos_ not in ["PUNCT", "PROPN"]]
```

```python
counts = [(word, words.count(word) / len(words)) for word in set(words)]
```

```python
word_freqs = [line.strip().split() for line in open("words_freqs.txt", "r").readlines() if len(line.split()) == 2]
```

```python
word_freqs = [(word, int(freq)) for word, freq in word_freqs]
sum_total = sum(freq for word, freq in word_freqs)
word_freqs = [(word, freq/sum_total) for word,freq in word_freqs]
```

```python
counts = {k:v for k,v in counts}
```

```python
word_freqs = {k:v for k,v in word_freqs}
```

```python
counts_compared = {k:counts[k]/word_freqs[k] for k in counts.keys() if k in word_freqs.keys() and words.count(k) > 10}
```

```python
sorted(counts_compared.items(), key=lambda r : r[-1], reverse=True)[2:50]
```
