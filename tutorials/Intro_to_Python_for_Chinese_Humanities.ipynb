```python
print("現在溫度：", temperature)
print("溫度是否高於35度？")
print(temperature > 35)

# let's change the temperature now. It is getting hotter!
temperature = 38.4
print("現在溫度：", temperature)
print("溫度是否高於35度？")
print(temperature > 35)

# it is raining, and temperature suddenly drops by 10 degrees
temperature = temperature - 10
print("現在溫度：", temperature)
print("溫度是否高於35度？")
print(temperature > 35)

# after the rain, the sun has come out. It is warmer again
temperature += 7
print("現在溫度：", temperature)
print("溫度是否高於35度？")
print(temperature > 35)
```

```python
text = "我今天很開心"
# for the computer, this is almost like
# text = ["我","今","天","很","開","心"]
# a list of characters

print("text有", len(text), "個漢字")
```

```python
i = 0
for character in text:
  print(i, character)
  i = i + 1
```

```python
for i, character in enumerate(text):
  print(i, character)
```

```python
print(text[3]) # print the character at index 3 (the 4th character; indexing starts from 0)
```

```python
words = ["我","今天","很","開心"]
print("text有", len(words), "個詞")
```

```python
for i, word in enumerate(words):
  print(i, word)
```
